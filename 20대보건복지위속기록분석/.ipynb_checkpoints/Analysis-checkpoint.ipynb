## Primitive Functions

```python
def gcol() : 
    stop = False
    i = 0
    while stop == False :
        i = i+1
        print('Collecting %d ...' % i)
        n = gc.collect()
        print('Unreachable objects :', n)
        print('Remaining Garbage :', gc.garbage)
        if n == 0 :
            stop = True
            
class prpr_data:
    def __init__(self, data) :
        self.data = data
    
    def by_party(self, party) : # Input must be lists!
        idx1 = self.data['PARTY'].isin(party)
        filtered_data = self.data[idx1]
        return prpr_data(filtered_data)

    def by_speaker(self, speaker) : # Input must be lists!
        idx2 = self.data['SPEAKER'].isin(speaker)
        filtered_data = self.data[idx2]
        return prpr_data(filtered_data)

    def by_meeting(self, meeting) : # Input must be lists!
        idx3 = self.data['MEETING_NAME'].isin(meeting)
        filtered_data = self.data[idx3]
        return prpr_data(filtered_data)
    
    def get_data(self) :
        return self.data

# prpr_data(HWC).by_speaker(['윤소하 의원']).get_data().head(3)
```

### 3.1. Main

```python
class Corpus:
    def __init__(self, docum, dictionary):
        self.docum = docum
        self.dictionary = dictionary
        self.length = 0
    def __iter__(self):
        for d in self.docum:
            yield self.dictionary.doc2bow(d)
    def __len__(self):
        if self.length == 0:
            for i, doc in enumerate(self.docum):
                continue
            self.length = i + 1
        return self.length

def get_topic_term_prob(lda_model):
    topic_term_freqs = lda_model.state.get_lambda()
    topic_term_prob = topic_term_freqs / topic_term_freqs.sum(axis=1)[:, None]
    return topic_term_prob

def doLDA(documents, ntopics, html_title, vis) :
    dictionary = gs.corpora.Dictionary(documents)
#     print('dictionary size : %d' % len(dictionary)) # dictionary size

#     Removing infrequent words
    min_count = 5
    word_counter = Counter((word for words in documents for word in words))
    removal_word_idxs = {
        dictionary.token2id[word] for word, count in word_counter.items()
        if count < min_count
    }

    dictionary.filter_tokens(removal_word_idxs)
    dictionary.compactify()
    # print('dictionary size : %d' % len(dictionary)) # dictionary size : 2652

#     Chunking
    corpus = Corpus(documents, dictionary)
#     for i, doc in enumerate(corpus): # testing corpus
#         if i==3 : 
#             print(doc)
#             break

#     Fitting
    lda_model = gs.models.LdaModel(corpus, id2word=dictionary, num_topics=ntopics)
#     print(lda_model.alpha.shape) # (n_topics,)
#     print(lda_model.alpha.sum()) # 1.0
    logperpl = lda_model.log_perplexity(corpus)
    coher = gs.models.CoherenceModel(lda_model, texts=documents, dictionary=dictionary, topn=10).get_coherence()

#     Checking
#     topic_term_prob = get_topic_term_prob(lda_model)
#     print('topics shape : ' + str(topic_term_prob.shape))     # (n_topics, n_terms)
#     print('topic probs : ' + str(topic_term_prob[0].sum()))  # 1.0

#     Visualization with pyLDAvis
    if vis == True :
        prepared_data = gsvis.prepare(lda_model, corpus, dictionary) # Take long time
        pyLDAvis.display(prepared_data)
        pyLDAvis.save_html(prepared_data, './result/' + html_title + str(ntopics) + '_pyLDAvis.html')

    return(lda_model, logperpl, coher)
```

### 3.2 Tuning Process: Train Perplexities & Coherence

```python
a, b = 1,100
ALL_perpl, ALL_coher = [], []
for top in tqdm(range(a,b+1)) :
    _,p,c = doLDA(noun_tokened['token'], top, 'ALL', False)
    ALL_perpl.append(p)
    ALL_coher.append(c)
    del _,p,c
```

```python
_, ax = plt.subplots(figsize=(9, 7))
ax.plot(range(a,b+1), ALL_perpl, '-.', label='ALL')
plt.title('Perplexities')
plt.xlabel('n Topics')
plt.ylabel('log perplexities')
ax.legend(fontsize=14)
plt.grid(True)
print('ALL ;',range(a,b+1)[np.argmin(ALL_perpl)])
```

```python
_, ax = plt.subplots(figsize=(9, 7))
ax.plot(range(a,b+1), ALL_coher, '-.', label='ALL')
plt.title('Coherences')
plt.xlabel('n Topics')
plt.ylabel('Coherence')
ax.legend(fontsize=14)
plt.grid(True)
print('ALL ;',range(a,b+1)[np.argmax(ALL_coher)])
```

```python
gcol()
# norm_corp = 보수['token'].copy()
norm_corp = 진보['token'].copy()
for j in tqdm(range(len(norm_corp))):
    norm_corp.iloc[j] = ' '.join(norm_corp.iloc[j])
norm_corp = norm_corp.tolist() # Normalized corpus.

print('Normalized corpus example : %s' %norm_corp[1:3])

cv = CountVectorizer(ngram_range=(1,1))
X = cv.fit_transform(norm_corp)
X
```

```python
Xc = (X.T * X) # This is the matrix manipulation step
Xc.setdiag(1) # We set the diagonals to be zeroes as it's pointless to be 1
names = cv.get_feature_names() # This are the entity names (i.e. keywords)
df = pd.DataFrame(data = Xc.toarray(), columns = names, index = names)
# df.to_csv('./result/보수coocurr_for_gephi2.csv', sep = ',', encoding = 'utf-8')
df.to_csv('./result/진보coocurr_for_gephi2.csv', sep = ',', encoding = 'utf-8')
```
