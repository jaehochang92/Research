# Monte Carlo Methods

## Motivation

For most probabilistic models of practical interest, exact inference is intractable, and so we have to resort to some form of approximation.

For a messy function $h:\mathbb{R}^p\rightarrow\mathbb{R}$ and a multivariate random variable ${\bf X}\in\mathbb{R}^p$ from a multivariate probability density $f:\mathbb{R}^p\rightarrow[0,\infty)$, approximate 

$$E[h({\bf X})]=\int h({\bf x})f({\bf x})d{\bf x}.$$

Typical numerical approximation involves the finite sum of function values over a grid of points. e.g., for $h:(a,b)\rightarrow\mathbb{R}$,

$$E[h(X_1)]\approx\sum_{i=1}^n\frac{h(X_i)}{n}:=\hat\mu_n$$

for a set of random samples $X_1,\cdots,X_n$ from $U(a,b)$. If $h$ is continuous and $n\rightarrow\infty$, $\hat\mu_n\overset{a.s.}{\rightarrow}E[h(X_1)]$ by the SLLN.

* If $p=1$, $E[h(X_1)]-\hat\mu_n=o_p(1)$.
* If $p>1$, we need $n^p$ points ${\bf X}_1,\cdots,{\bf X}_n$ to get an error of $o_p(1)$.

### Exercise

Integrate $$h(x)=\exp\{-\exp(\tan x)\}$$ over $x\in(-1.5, 1.5)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from sympy import *
%matplotlib inline

x, y = symbols('x y')
h = lambda x: exp(-exp(tan(x)))
plot_implicit(y < h(x), (x, -1.5, 1.5), (y, 0, 1), ylabel = '$h(x)$')
```

```python
sample_sizes = np.arange(1000, 10001, 1000)
mc = []
h_vec = np.vectorize(h)
for n in sample_sizes:
    random_samples = np.random.uniform(-1.5, 1.5, n)
    h_values = h_vec(random_samples)
    mc.append(sum(h_values)/n)
```

```python
for ch in ['ro', 'b--']:
    plt.plot(sample_sizes, mc, ch)
    plt.xlabel('grids')
    plt.ylabel('$\hat \int h$', rotation = 0) 
plt.hlines(mc[-1], min(sample_sizes), max(sample_sizes), 'red')
```
