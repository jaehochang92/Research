$\chi^2$ Independence Test: `Q5 ~ DM 1, 2, 4, 7, 10, 13, 14, 15`

```r
names(d.2020)
```

```r
y <- d.2020$Q5
X <- d.2020 %>% select(DM1, DM2, DM4, DM7, DM10, DM13:DM15)
titles <- names(X)

for (title in titles) {
    x <- X[, title, drop = T] %>% as_factor %>% gsub('~', '-', .)
    table.xy <- table(y, x %>% as_factor) %>% t
    t.obj <- chisq.test(table.xy)
    t.obj <- c(p.value = t.obj$p.value, t.obj$statistic)
    table.xy <- rbind(table.xy, colSums(table.xy)) %>% cbind(., rowSums(.))
    stargazer(table.xy, t.obj, out = fn$identity('`title`.html'), 
              type = 'text', summary = F, rownames = T)
}
```
