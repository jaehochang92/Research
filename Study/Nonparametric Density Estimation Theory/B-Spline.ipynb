```python
from sympy import symbols, integrate
from sympy.plotting import plot as splot
import matplotlib.pylab as plt
import numpy as np
```

```python
x,i = symbols('x,i')

# B-spline of order k
# i = 0,1,2,...
# k = 1,2,3,...

def B(x,t,i,k) :
    if k ==1 :
        return(1*(t[i]<=x)*(x<t[i+1]))
    else :
        return(
        (x-t[i])/(t[i+k-1]-t[i])*B(x,t,i,k-1)+(t[i+k]-x)/(t[i+k]-t[i+1])*B(x,t,i+1,k-1)
        )

t=[i for i in range(0,6,1)]
print(t)

i=1
for k in range(1,len(t)-i) :
    splot(B(x,t,i,k),(x,min(t),max(t)))
```
